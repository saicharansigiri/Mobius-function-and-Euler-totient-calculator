```python
import sympy
```

```python
 def numfac(n):
    count = 0
    a = []
    p = []
    for x in range (1,n+1):
        if n%x == 0:
            a.append(x)
            if sympy.isprime(x):
                p.append(x)
            
    return a,p;
def perfectfsquare(n):
    if (n**.5)%1 == 0 and n!=1:
        return True
```

```python
def mymain(val):
    val  = int(val)
    containsperfect = False
    a,p = numfac(val)
    for i in a:
        if perfectfsquare(i):
            containsperfect = True
    if val == 1:
        return 1
    elif len(p)%2 == 0 :
        return 1
    elif len(p)%2 !=0 and containsperfect :
        return 0
    elif containsperfect == False:
        return -1
```

```python
val  = input("Enter The number upto which you need mobius function: ")
val  = int(val)
print("Output:")
for i in range(1,val+1):
    print('\u03BC'+'('+str(i)+ ')'+' = ' + str(mymain(i)))
```
