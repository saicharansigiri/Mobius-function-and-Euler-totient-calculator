```python
    
def gcd(a, b): 
  
    if (a == 0): 
        return b 
    return gcd(b % a, a) 
  
def phi(n): 
  
    result = 1
    for i in range(2, n): 
        if (gcd(i, n) == 1): 
            result+=1
    return result 
```

```python
val = input("Enter The Number's upto which you need phi values: ")
val = int(val)
for n in range(1, val+1): 
    print("phi(",n,") = ", phi(n)) 
```
